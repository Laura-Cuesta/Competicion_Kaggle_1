### 2. Dividir X_train, X_test, y_train, y_test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[['Cpu','Ram']], y, test_size = 0.20, random_state = 42)
```

```python
X_train
```

### 3. Asignar el modelo (vacío) a una variable


```python
#VOY A ENTRENAR AHORA UN RADONTREEREGRESION
```

```python
from sklearn.ensemble import RandomForestRegressor

tree_reg = RandomForestRegressor(max_depth=3, random_state=42, n_estimators=150)
tree_reg.fit(X_train, y_train)
```

### 4. Sacar métricas, valorar el modelo 

Recuerda que en la competición se va a evaluar con la métrica de MAE.

```python
prediciones_tre_reg_train = tree_reg.predict(X_train)
prediciones_tre_reg_test = tree_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.metrics import mean_squared_error
```

```python

#4. SACO LAS METRICAS DE ERROR ENTRE LOS TRAIN Y LOS TEST PARA VER COMO HA CAMBIADO
print('Train MAE:', mean_absolute_error(y_train, prediciones_tre_reg_train))
print('Test MAE:', mean_absolute_error(y_test, prediciones_tre_reg_test),'\n')
```
